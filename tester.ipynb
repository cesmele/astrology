```python
class Person:
    
    def __init__(self, name, zodiac, chinese_zodiac):
        
        self.name = name
        self.zodiac = zodiac
        self.chinese_zodiac = chinese_zodiac
        
    def __rep__(self):
      return f"Person({self.name!r}, {self.zodiac!r}, {self.chinese_zodiac!r})"
        
    
class People:
    
    def __init__(self, path):
        
        self.catalog = {}
        with open(path, "r", encoding = "utf-8") as f:
            for line in f:
                line = line.strip()
                split_line = line.split(",")
                name = split_line[0]
                birthday = split_line[1]
                self.catalog[name] = birthday
                
    def create_profile(self, name):
        
        if name in self.catalog:
            birthday = self.catalog.get(name)
            birth_data_file = date_iterator(birthday)
            month = int(birth_data_file[0])
            day = int(birth_data_file[1])
            year = int(birth_data_file[2])
            zodiac = zodiac_convert(month, day)
            chinese_zodiac = chinese_zodiac_convert(year)
            return Person(name, zodiac, chinese_zodiac)
        else:
            raise KeyError("Name not found")
          
          
def date_iterator(date):
    
    date.split("/")
    date_split = date.split("/")
    month = date_split[0]
    day = date_split[1]
    year = date_split[2]
                
    return month, day, year
        
def compatibilities(zodiacsign):

    compatibility = ()
    if zodiacsign == "Aries":
        compatibility.append("Sagittarius", "Gemini", "Libra")
        return compatibility
    elif zodiacsign == "Taurus":
        compatibility.append("Capricon", "Virgo", "Cancer")
        return compatibility
    elif zodiacsign == "Gemini": 
        compatibility.append("Libra", "Sagittarius", "Aquarius")
        return compatibility
    elif zodiacsign == "Cancer": 
        compatibility.append("Scorpio", "Pisces", "Virgo")
        return compatibility
    elif zodiacsign == "Leo":
        compatibility.append("Sagittarius", "Aquarius", "Gemini")
        return compatibility
    elif zodiacsign == "Virgo":
        compatibility.append("Scorpio", "Taurus", "Capricorn")
        return compatibility
    elif zodiacsign == "Libra":
        compatibility.append("Libra", "Aries", "Aquarius")
        return compatibility
    elif zodiacsign == "Scorpio":
        compatibility.append("Cancer", "Pisces", "Virgo")
        return compatibility
    elif zodiacsign == "Capricorn":
        compatibility.append("Virgo", "Capricorn")
        return compatibility
    elif zodiacsign == "Aquarius":
        compatibility.append("Libra", "Gemini", "Aquarius")
        return compatibility
    elif zodiacsign == "Pisces":
        compatibility.append("Cancer", "Capricorn", "Scorpio")
        return compatibility


def zodiac_convert(month, day):

  if month == 1:
    if day <= 19:
      return "Capricorn"
    else:
      return "Aquarius"
  elif month == 2:
    if day <= 18:
      return "Aquarius"
    else:
      return "Pisces"
  elif month == 3:
    if day <= 20:
      return "Pisces"
    else:
      return "Aries"
  elif month == 4:
    if day <= 19:
      return "Aries"
    else:
      return "Taurus"
  elif month == 5:
    if day <= 20:
      return "Taurus"
    else:
      return "Gemini"
  elif month == 6:
    if day <= 20:
      return "Gemini"
    else:
      return "Cancer"
  elif month == 7:
    if day <= 22:
      return "Cancer"
    else:
      return "Leo"
  elif month == 8:
    if day <= 22:
      return "Leo"
    else:
      return "Virgo"
  elif month == 9:
    if day <= 22:
      return "Virgo"
    else:
      return "Libra"
  elif month == 10:
    if day <= 22:
      return "Libra"
    else:
      return "Scorpio"
  elif month == 11:
    if day <= 21:
      return "Scorpio"
    else:
      return "Sagittarius"
  elif month == 12:
    if day <= 21:
      return "Sagittarius"
    else:
      return "Capricorn"
                
def chinese_zodiac_convert(year):
        zodiac={0:"Monkey",
                1:"Rooster",
                2:"Dog",
                3:"Pig",
                4:"Rat",
                5:"Ox",
                6:"Tiger",
                7:"Rabbit",
                8:"Dragon",
                9:"Snake",
                10:"Horse",
                11:"Sheep"
                }
        key= int(year)%12
        return zodiac[key]
```
